Import Dependencies

```python
import pandas as pd
from sqlalchemy import create_engine
from config import password
```

Use Pandas to read CSV File #1 & store as dataframe

```python
csv_file = "AB_NYC_2019.csv"
NYC_Data_df = pd.read_csv(csv_file)
NYC_Data_df.head()
```

Inspect the data above and edit/transform accordingly 

```python
#rename column headers for cleaner comparison of data (NYC vs. Boston) 
NYC_Data_df.rename(columns={'id': 'listing_id', 'name': 'summary', 'last_review': 'date'}, inplace=True)
NYC_Data_df.head()
```

```python
#show only pertinent information for comparison to Boston data 
Cleaned_NYC_data_df = NYC_Data_df[['listing_id', 'reviews_per_month', 'summary', 'date', 'price']].copy()
Cleaned_NYC_data_df.head()
```

Use Pandas to read CSV File #2

```python
csv_file = "calendar.csv"
Boston_Cal_df = pd.read_csv(csv_file)
Boston_Cal_df.head()
```

Use Pandas to read CSV File #3

```python
csv_file = "listings.csv"
Boston_Listing_df = pd.read_csv(csv_file)
Boston_Listing_df.head()
```

```python
#rename 'id' to 'listing id' for cleaner table comparison 
Boston_Listing_df.rename(columns={'id': 'listing_id'}, inplace=True)
Boston_Listing_df.head()
```

Use Pandas to read CSV File #4

```python
csv_file = "reviews.csv"
Boston_Rev_df = pd.read_csv(csv_file)
Boston_Rev_df.head()
```

```python
#drop 'id' column - irrelevant, we will use listing_id for comparison
Boston_Rev_df.drop('id', axis=1, inplace=True)
Boston_Rev_df.head()
```

```python
# Merge Boston data (dataframes) into a single dataframe - Part 1
Boston_Data_df = pd.merge(Boston_Listing_df,Boston_Rev_df, how="inner")
Boston_Data_df.head(4)
```

```python
# Merge Boston data (dataframes) into a single dataframe - Part 2
New_Boston_Data_df = pd.merge(Boston_Data_df,Boston_Cal_df, how="inner")
New_Boston_Data_df.head(4)
```

```python
#show only pertinent information - filter data
Cleaned_Boston_data_df = Boston_Data_df[['listing_id', 'reviews_per_month', 'summary', 'date', 'price']].copy()
Cleaned_Boston_data_df.head()
```

Connect Local Database

```python
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/AirBNB_db')
connection = engine.connect()
```

```python
engine.table_names()
```

```python
Cleaned_NYC_data_df.to_sql(name='nyc_airbnb_data', con=engine, if_exists='append', index=False)
```

```python
Cleaned_Boston_data_df.to_sql(name='boston_airbnb_data', con=engine, if_exists='append', index=False)
```

```python
#Confirm data has been added by querying the nyc_airbnb_data table
pd.read_sql_query('select * from nyc_airbnb_data', con=engine).head()
```

```python
#Confirm data has been added by querying the boston_airbnb_data table
pd.read_sql_query('select * from boston_airbnb_data', con=engine).head()
```
